### Проверка правильности расчета эффективности обогащения

Сохраним значения столбцов в переменные и вычислим эффективность обогащения по формуле:

$Recovery = {{C*(F-T)}\over{F*(C-T)}}*100%$

```python
# создадим три переменные, чтобы не строить километровый рассчет
c = df_train['rougher.output.concentrate_au']
f = df_train['rougher.input.feed_au']
t = df_train['rougher.output.tail_au']
# считаем
df_train['rougher.output.recovery.check'] = (c * (f - t)) / (f * (c - t)) * 100
```

Рассчитаем среднюю абсолютную ошибку

```python
print('MAE для расчетов эффективности обогащения равна', 
      mean_absolute_error(df_train['rougher.output.recovery'], df_train['rougher.output.recovery.check']))
```

Средняя абсолютная ошибка равна $9.7 * 10^{-15}$. Это очень маленькое среднее абсолютных разностей сравниваемых значений. Исходя из этого, будем считать, что значения `rougher.output.recovery` рассчитаны корректно.

```python
# удалим вновь созданный столбец за ненадобностью
df_train = df_train.drop('rougher.output.recovery.check', axis=1)
```

### Вводим метрику sMAPE
Создадим функцию, для определения sMAPE для каждой целевой перемнной по следующей формуле:

$sMAPE = {1\over{N}}{\sum\limits_{i=1}^N}{{|y_i - \hat{y_i}|}\over{(|y_i| + |\hat{y_i}|)/2}}*100%$

Затем введем итоговое sMAPE и сделаем его метрикой.

```python
def sMAPE(y_true, y_predict):
    smape = ((abs(y_true - y_predict) / ((abs(y_true) + abs(y_predict))/2)).mean()) * 100
    final_smape = 0.25*smape[0] + 0.75*smape[1]
    return final_smape
```

Вводим метрику

```python
smape_scorer = make_scorer(sMAPE, greater_is_better=False)
```
